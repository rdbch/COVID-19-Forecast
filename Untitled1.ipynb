```python
import os
import pandas as pd
from src.core.data import CountryDataset
from src.core.networks import BasicRecurrentPredictor
from torch import nn
from torch import optim
import torch
import seaborn as sns
sns.set()
```

```python
countryData = CountryDataset('assets', windowSize = 10, predSize = 5, batchSize = 16, mode = 'confirmed',
                        state = 'Poland', topSimilar = 5, alignThreshConf = 300, alignThreshDead = 10,
                        device = 'cuda:0', trainUpTo = pd.to_datetime('2020-04-07'))
```

```python
confTrain = torch.from_numpy(countryData.confTrain).float().cuda()
deadTrain = torch.from_numpy(countryData.deadTrain).float().cuda()
```

```python
# build the model
seq = BasicRecurrentPredictor(chNo = 1, hidChNo = 128, future = 0, teacherProb = 0.5,
            # rnn cell type and depth
            rnnCell    = 'LSTMCell',
            rnnNoCells = 2,

            # multi layer perceptron that is applied to the output of rnn
            mlpLayerCfg   = [64, 64],    # layer hidden dims
            mlpActiv      = 'PReLU',     # inner activation of the mlp
            dropRate      = None,       # dropout rate for each layer of mlp
            normType      = None,
            mlpActivLast  = None,       # note that every timestamp in the sequence will be activated too
            returnFullSeq = True)
seq.build()
seq.cuda()
criterion = nn.MSELoss()

# use LBFGS as optimizer since we can load the whole data to train
optimizer = optim.LBFGS(seq.parameters(), lr=0.03, max_iter=500, tolerance_grad=1e-6, tolerance_change=1e-9)
lrSched   = optim.lr_scheduler.StepLR(optimizer, 1, gamma=0.9997, last_epoch=-1)
#begin to train
for i in range(10):
    print('STEP: ', i)
    
    def closure():
        optimizer.zero_grad()
        out  = seq(confTrain).reshape(-1, 10, 1)
        loss = criterion(out, confTrain)
        print('loss:', loss.item(), optimizer.param_groups[0]['lr'])
        loss.backward()
        lrSched.step()
        return loss
    
    optimizer.step(closure)
```

```python
sns.lineplot(x = [i for i in range(10)],y = confTrain[100].squeeze(-1).cpu())
res = seq(confTrain[100].unsqueeze(0), future = 100).cpu().detach().numpy()
print(res.shape)
sns.lineplot(x = [i for i in range(110)],y = res.squeeze(0))
# confTrain[100].squeeze(-1)
```
